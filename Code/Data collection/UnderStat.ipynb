```python
import pandas as pd
import os
league_list = ['epl', 'laliga', 'serieA', 'bundesliga', 'ligue1']
```

```python
documents = os.path.join(os.path.expanduser("~"), "Documents")

for league in league_list:
    team_data = pd.read_csv(f'C:/Users/ASUS/Downloads/{league}-league-chemp.csv', sep=';')
    filename = f"{league}_understat_24-25.csv"
    team_data.to_csv(os.path.join(documents, filename), index=False)
    print(f"Saved {league} data.")
```
