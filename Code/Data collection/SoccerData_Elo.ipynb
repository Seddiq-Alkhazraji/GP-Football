```python
print("Initializing ClubElo scraper...")
elo = sd.ClubElo()

# Define the output folder
output_dir = os.path.join(os.path.expanduser("~"), "Documents")

print(f"ðŸ“‚ Saving files to: {output_dir}")

# --- 1. Fetch 2023/24 Season (Historical Data) ---
print("\n Fetching Aug 15, 2023 (Start of 23/24)...")
elo_rank = elo.read_by_date(date='2023-08-15')
elo_start_23 = elo_rank.reset_index()
elo_start_23.to_csv(os.path.join(output_dir, "elo_start_23.csv"), index=False)
print("   Saved elo_start_23.csv")

print("  Fetching Jan 15, 2024 (Mid 23/24)...")
elo_rank = elo.read_by_date(date='2024-01-15')
elo_end_23 = elo_rank.reset_index()
elo_end_23.to_csv(os.path.join(output_dir, "elo_end_23.csv"), index=False)
print("   Saved elo_end_23.csv")

# --- 2. Fetch 2024/25 Season (Modern Market Data) ---
print("\n  Fetching Aug 15, 2024 (Start of 24/25)...")
elo_rank = elo.read_by_date(date='2024-08-15')
elo_start_24 = elo_rank.reset_index()
elo_start_24.to_csv(os.path.join(output_dir, "elo_start_24.csv"), index=False)
print("   Saved elo_start_24.csv")

print("  Fetching Jan 01, 2025 (Mid 24/25)...")
elo_rank = elo.read_by_date(date='2025-01-01')
elo_end_24 = elo_rank.reset_index() 
elo_end_24.to_csv(os.path.join(output_dir, "elo_end_24.csv"), index=False)
print("   Saved elo_end_24.csv")

print("\n All 4 Elo files are ready!")
```
